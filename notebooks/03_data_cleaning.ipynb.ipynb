```python
import pandas as pd

gdp = pd.read_csv(
    r"C:\Users\jacob\Desktop\Finscope-main\data\raw\gdp_per_capita_ppp.csv"
)

gdp_clean = gdp[[
    "REF_AREA_LABEL",
    "TIME_PERIOD",
    "OBS_VALUE"
]].copy()

gdp_clean.columns = [
    "Country",
    "Year",
    "GDP_Per_Capita_PPP"
]

gdp_clean.head()
```

```python
inflation = pd.read_csv(
    r"C:\Users\jacob\Desktop\Finscope-main\data\raw\inflation.csv",
    skiprows=4
)

inflation_long = inflation.melt(
    id_vars=["Country Name"],
    var_name="Year",
    value_name="Inflation"
)

inflation_long.rename(
    columns={"Country Name":"Country"},
    inplace=True
)

inflation_long.head()
```

```python
unemployment = pd.read_csv(
    r"C:\Users\jacob\Desktop\Finscope-main\data\raw\unemployment.csv",
    skiprows=4
)

unemployment_long = unemployment.melt(
    id_vars=["Country Name"],
    var_name="Year",
    value_name="Unemployment"
)

unemployment_long.rename(
    columns={"Country Name":"Country"},
    inplace=True
)

unemployment_long.head()
```

```python
gdp_clean["Year"] = gdp_clean["Year"].astype(str)

inflation_long["Year"] = inflation_long["Year"].astype(str)

unemployment_long["Year"] = unemployment_long["Year"].astype(str)
```

```python
merged = pd.merge(
    gdp_clean,
    inflation_long,
    on=["Country","Year"],
    how="inner"
)

merged = pd.merge(
    merged,
    unemployment_long,
    on=["Country","Year"],
    how="inner"
)

print(merged.shape)

merged.head()
```

```python
# Keep only years from 2010 onwards

merged["Year"] = pd.to_numeric(merged["Year"], errors="coerce")

merged = merged[
    (merged["Year"] >= 2010) &
    (merged["Year"] <= 2025)
]

print("Shape after year filter:", merged.shape)

merged.head()
```

```python
merged = merged.reset_index(drop=True)
```

```python
merged.head()
```

```python
merged = merged[
    (merged["Year"] >= 2010) &
    (merged["Year"] <= 2025)
]

merged = merged.reset_index(drop=True)

print(merged.shape)
merged.head()
```

```python
print(merged["Year"].min())
print(merged["Year"].max())
```

```python
# Remove rows with missing values

merged = merged.dropna()

print("Shape after removing nulls:", merged.shape)
```
